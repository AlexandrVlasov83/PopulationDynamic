```python
import pandas as pd
import numpy as np
import re
```

```python
city_info = pd.read_html('https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%B3%D0%BE%D1%80%D0%BE%D0%B4%D0%BE%D0%B2_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8')
```

```python
city_info[0].columns = ['number','coat_of_arms','city','region','federal_district','population_new','first_year','city_year','old_name']
city_info[1].columns = ['number','coat_of_arms','city','region','federal_district','population_new','first_year','city_year','old_name']
```

```python
city_info[0]
```

```python
city_info[1]
```

```python
city_info_all = pd.concat([city_info[0],city_info[1]],sort=False, axis=0, ignore_index=True)[['city','region','federal_district','population_new']]
```

```python
city_info_all.tail()
```

```python
def num_cor(data):
    return re.sub(' ','',data)
```

```python
def num_pop(data):
    return re.search(r'\d+',data).group()
```

```python
city_info_all['population_new'] = city_info_all.population_new.apply(num_cor)
city_info_all['population_new'] = city_info_all.population_new.apply(num_pop)
city_info_all = city_info_all.sort_values('city', ascending=True)
```

```python
city_info_all.tail()
```

```python
city_2017 = city_2017.dropna()
```

```python
city_2017.tail()
```

```python
def city_name(data):
    name = re.match(r'(?P<g>[г])(?P<dot>\.)(?P<name>\D+)',data)
    return name.group('name')
```

```python
city_2017['city'] = city_2017.city.apply(city_name)
```

```python
city_2017 = city_2017.sort_values('city', ascending=True, ignore_index=True)[['city','population']]
```
